```python
# Importar librerías y módulos
import numpy as np
from sklearn.model_selection import cross_val_predict, GridSearchCV
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
```

## Ejercicio 2

**2 - M_2.txt**


```python
# Importar el conjunto de datos
ma_a = np.loadtxt('/content/drive/MyDrive/z. Aprendizaje/M_2.txt')
print('Observaciones:', ma_a.shape[0])
print('Variables:', ma_a.shape[1]-2)
```

## 1. Evalúa un modelo perceptrón multicapa con validación cruzada utilizando al menos 5 capas de 20 neuronas.

```python
# MODELO PERCEPTRÓN MULTICAPA
print('MODELO PERCEPTRÓN MULTICAPA')

# Matriz de las variables predictoras
x = ma_a[:, 2:]

# Arreglo de la variable de respuesta
y = ma_a[:, 0]

# Evaluar el modelo perceptrón con validación cruzada y 5 capas de 20 neuronas
y_pred = cross_val_predict(MLPClassifier(hidden_layer_sizes=(20, 20, 20, 20, 20), max_iter=50000), x, y)
print(classification_report(y, y_pred))
```

## 2. Evalúa un modelo perceptrón multicapa con validación cruzada, pero encontrando el número óptimo de capas y neuronas de la red.


```python
# ENCONTRAR NÚMERO ÓPTIMO DE CAPAS

# Arreglo de valores propuestos para el número de capas y neuronas
num_layers = np.arange(1, 20, 5)
num_neurons = np.arange(10, 110, 20)

# Arreglo para guardar las exactitudes de cada capa
layers = []

# Por cada capa propuesta
for l in num_layers:
    # Por cada neurona propuesta
    for n in num_neurons:
        # Crear arreglo arquitectura de la red y a;adirlo al arreglo correspondiente
        layers.append(l*[n])

# Calcular los hiperparámetros óptimos para el modelo
clf = GridSearchCV(MLPClassifier(max_iter=10000), {'hidden_layer_sizes': layers}, cv=5)
clf.fit(x, y)
print(clf.best_estimator_)
```

```python
# EVALUAR CON VALIDACIÓN CRUZADA
y_pred = cross_val_predict(clf, x, y, cv=5)
print(classification_report(y, y_pred))
```

## 3. Prepara el modelo perceptrón multicapa:
### A. Opten los hiperparámetros óptimos de capas y neuronas de la red.
### B. Con los hiperparámetros óptimos, ajusta el modelo con todos los datos.


```python
# MLP CLASSIFIER
print('MLP CLASSIFIER')

# Hiperparámetros óptimos
clf = GridSearchCV(MLPClassifier(max_iter=10000), {'hidden_layer_sizes': layers}, cv=5)
y_pred = cross_val_predict(clf, x, y, cv=5)

# Ajuste del modelo con todos los datos
clf.fit(x, y)

# Impresión de resultados
print(clf.best_estimator_)
print(classification_report(y, y_pred))
```
