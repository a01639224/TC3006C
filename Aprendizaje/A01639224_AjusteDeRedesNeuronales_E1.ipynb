```python
# Importar librerías y módulos
import pandas as pd
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.model_selection import KFold
import numpy as np
```

## Ejercicio 1

**4 - Variable dependiente MR, variables independientes M, W, H y P**


```python
# Importar el conjunto de datos
cd_df = pd.read_csv('/content/drive/MyDrive/z. Aprendizaje/crime_data.csv')
cd_df.head()
```

## 1. Evalúa con validación cruzada un modelo pereceptrón multicapa para las variables que se te asignaron para este ejercicio.

```python
# MODELO PERCEPTRÓN MULTICAPA
print('MODELO PERCEPTRÓN MULTICAPA')

# Matriz de las variables predictoras
x = cd_df.drop(['State', 'VR',	'MR',	'S'], axis=1).to_numpy()

# Arreglo de la variable de respuesta
y = cd_df['MR'].to_numpy()

# Fit regression model using the complete dataset and calculate MSE and MAE
regr = MLPRegressor(hidden_layer_sizes=(20, 20), max_iter=20000)
# hidden_layer_sizes : arquitectura de la red (n capas cada una con x neuronas)
#max_iter : número máximo de iteraciones (epochs) durante el entrenamiento
regr.fit(x, y)

# Predecir variable predictora
y_pred = regr.predict(x)

# Calcular MSE y MAE
print('MSE:', mean_squared_error(y, y_pred))
print('MAE:', mean_absolute_error(y, y_pred))
```

```python
# EVALUACIÓN CON VALIDACIÓN CRUZADA
print('EVALUACIÓN CON VALIDACIÓN CRUZADA')

# Número de plieges
n_folds = 3

# Dividir los conjuntos de prueba y entrenamiento
kf = KFold(n_splits=n_folds, shuffle = True)

# Arreglo para guardar los resultados de las pruebas
mse_cv = []
mae_cv = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x):

    # Definir el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]

    # Entrenar el modelo de regresión con los datos de entrenamiento
    regr.fit(x_train, y_train)

    # Definir el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]

    # Predecir las variables predictoras del conjunto de prueba
    y_pred = regr.predict(x_test)

    # Calcular los valores de MSE, MAE y añadir a sus respectivos arreglos
    mse_i = mean_squared_error(y_test, y_pred)
    mse_cv.append(mse_i)
    mae_i = mean_absolute_error(y_test, y_pred)
    mae_cv.append(mae_i)

# Imprimir los resultados
print('MSE:', np.average(mse_cv))
print('MAE:', np.average(mae_cv))
```

## 2. Agrega al conjunto de datos columnas que representen los cuadrados de las variables predictoras (por ejemplo, M^2, W^2), así como los productos entre pares de variables (por ejemplo, PxS, MxW). Evalúa un modelo perceptrón multicapa para este nuevo conjunto de datos.


```python
# Añadir cuadrados y productos de variables predictoras
cd_df["M^2"] = cd_df['M']**2
cd_df["W^2"] = cd_df['W']**2
cd_df["H^2"] = cd_df['H']**2
cd_df["P^2"] = cd_df['P']**2
cd_df["MxW"] = cd_df['M']*cd_df['W']
cd_df["MxH"] = cd_df['M']*cd_df['H']
cd_df["MxP"] = cd_df['M']*cd_df['P']
cd_df["WxH"] = cd_df['W']*cd_df['H']
cd_df["WxP"] = cd_df['W']*cd_df['P']
cd_df["HxP"] = cd_df['H']*cd_df['P']
cd_df.head()
```

```python
# MODELO PERCEPTRÓN MULTICAPA
print('MODELO PERCEPTRÓN MULTICAPA')

# Matriz de las variables predictoras
x = cd_df.drop(['State', 'VR',	'MR',	'S'], axis=1).to_numpy()

# Arreglo de la variable de respuesta
y = cd_df['MR'].to_numpy()

# Fit regression model using the complete dataset and calculate MSE and MAE
regr.fit(x, y)

# Predecir variable predictora
y_pred = regr.predict(x)

# Calcular MSE y MAE
print('MSE:', mean_squared_error(y, y_pred))
print('MAE:', mean_absolute_error(y, y_pred))
```

```python
# EVALUACIÓN CON VALIDACIÓN CRUZADA
print('EVALUACIÓN CON VALIDACIÓN CRUZADA')

# Número de plieges
n_folds = 3

# Dividir los conjuntos de prueba y entrenamiento
kf = KFold(n_splits=n_folds, shuffle = True)

# Arreglo para guardar los resultados de las pruebas
mse_cv = []
mae_cv = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x):

    # Definir el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]

    # Entrenar el modelo de regresión con los datos de entrenamiento
    regr.fit(x_train, y_train)

    # Definir el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]

    # Predecir las variables predictoras del conjunto de prueba
    y_pred = regr.predict(x_test)

    # Calcular los valores de MSE, MAE y añadir a sus respectivos arreglos
    mse_i = mean_squared_error(y_test, y_pred)
    mse_cv.append(mse_i)
    mae_i = mean_absolute_error(y_test, y_pred)
    mae_cv.append(mae_i)

# Imprimir los resultados
print('MSE:', np.average(mse_cv))
print('MAE:', np.average(mae_cv))
```
