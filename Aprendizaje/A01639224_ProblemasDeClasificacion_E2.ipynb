## 2. Evalúa al menos 5 modelos de clasificación distintos utilizando validación cruzada, y determina cuál de ellos es el más efectivo.

```python
# Identificar variables de prueba y de respuesta
x = ma_a[:, 2:]
y = ma_a[:, 0]

# Número de plieges
n_folds = 5

# Dividir los conjuntos de prueba y entrenamiento
kf = KFold(n_splits=n_folds, shuffle=True)
```

```python
# Linear SVM
print('Linear SVM')

# Arreglo para evaluar el desempeño del modelo
cv_y_test = []
cv_y_pred = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x, y):

    # Entrenar la regresión con el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]
    clf = SVC(kernel='linear')
    clf.fit(x_train, y_train)

    # Probar la regresión con el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = clf.predict(x_test)

    # Almacenar los valores verdaderos y predicciones
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Evaluar el desempeño del modelo
print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))
```

```python
# RBF SVM
print('RBF SVM')

# Arreglo para evaluar el desempeño del modelo
cv_y_test = []
cv_y_pred = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x, y):

    # Entrenar la regresión con el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]
    clf = SVC(kernel='rbf')
    clf.fit(x_train, y_train)

    # Probar la regresión con el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = clf.predict(x_test)

    # Almacenar los valores verdaderos y predicciones
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Evaluar el desempeño del modelo
print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))
```

```python
# K-NEAREST NEIGHBORS
print('K-NEAREST NEIGHBORS')

# Arreglo para evaluar el desempeño del modelo
cv_y_test = []
cv_y_pred = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x, y):

    # Entrenar la regresión con el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]
    clf = KNeighborsClassifier(n_neighbors=3)
    clf.fit(x_train, y_train)

    # Probar la regresión con el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = clf.predict(x_test)

    # Almacenar los valores verdaderos y predicciones
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Evaluar el desempeño del modelo
print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))
```

```python
# DECISION TREE
print('DECISION TREE')

# Arreglo para evaluar el desempeño del modelo
cv_y_test = []
cv_y_pred = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x, y):

    # Entrenar la regresión con el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]
    clf = DecisionTreeClassifier()
    clf.fit(x_train, y_train)

    # Probar la regresión con el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = clf.predict(x_test)

    # Almacenar los valores verdaderos y predicciones
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Evaluar el desempeño del modelo
print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))
```

```python
# LINEAR DISCRIMINANT ANALYSIS
print('LINEAR DISCRIMINANT ANALYSIS')

# Arreglo para evaluar el desempeño del modelo
cv_y_test = []
cv_y_pred = []

# Repetir el proceso por cada pliege
for train_index, test_index in kf.split(x, y):

    # Entrenar la regresión con el conjunto de entrenamiento
    x_train = x[train_index, :]
    y_train = y[train_index]
    clf = LinearDiscriminantAnalysis()
    clf.fit(x_train, y_train)

    # Probar la regresión con el conjunto de prueba
    x_test = x[test_index, :]
    y_test = y[test_index]
    y_pred = clf.predict(x_test)

    # Almacenar los valores verdaderos y predicciones
    cv_y_test.append(y_test)
    cv_y_pred.append(y_pred)

# Evaluar el desempeño del modelo
print(classification_report(np.concatenate(cv_y_test), np.concatenate(cv_y_pred)))
```

El modelo de clasificación SVM linear es el más efectivo ya que genera los mejores resultados en la mayoría de los rubros del reporte de clasificación. Pese a que el modelo SVM RBF también es efectivo, el kernel linear produce mejores promedios en la mayoría de los resultados globales.

## 3. Escoge al menos dos clasificadores que hayas evaluado en el paso anterior e identifica sus hiperparámetros. Lleva a cabo el proceso de validación cruzada anidada para evaluar los dos modelos con la selección óptima de hiperparámetros.


Para el clasificador SVM con kernel linear  el hiperparámetro a calcularse es el de Parámetro de Regularización (C).

```python
# SVM LINEAR
print('SVM LINEAR')

# Arreglo de valores propuestos para C (Parametro de regularización)
cc = np.logspace(-4, 0, 100)

# Arreglo para guardar las exactitudes de cada C
acc = []

# Por cada C propuesta
for c in cc:

    # Arreglo para guardar las exactitudes de las pruebas
    acc_cv = []

    # Dividir los conjuntos de prueba y entrenamiento
    kf = KFold(n_splits=5, shuffle=True)

    # Repetir el proceso por cada pliege
    for train_index, test_index in kf.split(x, y):

        # Crear un modelo con la C indicada
        x_train = x[train_index, :]
        y_train = y[train_index]
        clf_cv = SVC(C=c, kernel='linear')

        # Entrenar el modelo con el conjunto de entrenamiento
        clf_cv.fit(x_train, y_train)

        # Probar el modelo con el conjunto de prueba
        x_test = x[test_index, :]
        y_test = y[test_index]
        y_pred = clf_cv.predict(x_test)

         # Calcular la exactitud del modelo y añadir al arreglo
        acc_i = accuracy_score(y_test, y_pred)
        acc_cv.append(acc_i)

    # Calcular la exactitud para cada C
    acc_hyp = np.average(acc_cv)
    acc.append(acc_hyp)

    # Imprimir el resultado de las pruebas para cada C
    print('C:', c, '| ACC:', acc_hyp)

# Imprimir el valor óptimo de C
opt_index = np.argmax(acc)
opt_hyperparameter = cc[opt_index]
print('Valor óptimo de C:', opt_hyperparameter)
```

Para el clasificador SVM con kernel RBF el hiperparámetro a calcularse es el Coeficiente del Kernel (gamma).

```python
# SVM RBF
print('SVM RBF')

# Arreglo de valores propuestos de gamma (Coeficiente del kernel)
gg = np.logspace(-5, -1, 100)

# Arreglo para guardar las exactitudes de cada gamma
acc = []

# Por cada gamma propuesto
for g in gg:

    # Arreglo para guardar las exactitudes de las pruebas
    acc_cv = []

    # Dividir los conjuntos de prueba y entrenamiento
    kf = KFold(n_splits=5, shuffle=True)

    # Repetir el proceso por cada pliege
    for train_index, test_index in kf.split(x, y):

        # Crear un modelo con la C indicada
        x_train = x[train_index, :]
        y_train = y[train_index]
        clf_cv = SVC(kernel ='rbf', gamma = g)

        # Entrenar el modelo con el conjunto de entrenamiento
        clf_cv.fit(x_train, y_train)

        # Probar el modelo con el conjunto de prueba
        x_test = x[test_index, :]
        y_test = y[test_index]
        y_pred = clf_cv.predict(x_test)

        # Calcular la exactitud del modelo y añadir al arreglo
        acc_i = accuracy_score(y_test, y_pred)
        acc_cv.append(acc_i)

    # Calcular la exactitud para cada gamma
    acc_hyp = np.average(acc_cv)
    acc.append(acc_hyp)

    # Imprimir el resultado de las pruebas para cada gamma
    print('Gamma:', g, '| ACC:', acc_hyp)

# Imprimir el valor óptimo de C
opt_index = np.argmax(acc)
opt_hyperparameter = gg[opt_index]
print('Valor óptimo de gamma:', opt_hyperparameter)
```

## 4. Prepara tus modelos para producción haciendo lo siguiente:

### A. Opten los hiperparámetros óptimos utilizando todo el conjunto de datos con validación cruzada.
### B. Con los hiperparámetros óptimos, ajusta el modelo con todos los datos.

```python
# SVM LINEAR
print('SVM LINEAR')

# Hiperparámetros óptimos
clf = GridSearchCV(SVC(kernel='linear'), {'C': np.logspace(-4, 0, 100)}, cv=5)
y_pred = cross_val_predict(clf, x, y, cv=5)

# Ajuste del modelo con todos los datos
clf.fit(x, y)

# Impresión de resultados
print(clf.best_estimator_)
print(classification_report(y, y_pred))
```

```python
# SVM RBF
print('SVM RBF')

# Hiperparámetros óptimos
clf = GridSearchCV(SVC(kernel='rbf'), {'gamma': np.logspace(-5, -1, 100)}, cv=5)
y_pred = cross_val_predict(clf, x, y, cv=5)

# Ajuste del modelo con todos los datos
clf.fit(x, y)

# Impresión de resultados
print(clf.best_estimator_)
print(classification_report(y, y_pred))
```
