# Importar los datos

```python
# Import necessary libraries and modules
import pandas as pd # Read the .csv file and turn it into a DataFrame
from sklearn.model_selection import train_test_split # Split the data into train and test sets
from sklearn.metrics import accuracy_score # Calculate the accuracy of the model
from sklearn.metrics import confusion_matrix # Measure the performance of the model
import seaborn as sns # Graph the confusion matrix
import matplotlib.pyplot as plt # Graph the confusion matrix
```

```python
# Read the .csv file and turn it into a DataFrame
iris_df = pd.read_csv('/content/drive/MyDrive/z. Estadística/Iris.csv')
```

```python
# Drop 'Id' column from DataFram
iris_df = iris_df.drop(['Id'], axis=1)
```

```python
# Split the data into train and test sets
x_train, x_test, y_train, y_test = train_test_split(iris_df.iloc[:,:-1], iris_df.iloc[:,-1], test_size=0.2, random_state=42)
# Feature matrix = exclude last column; Response vector = last column; Test size = 20% of data; Random state = Same split every time
```

```python
# Get the order of the unique response values in the training set (for confusion matrix labeling)
y_test.unique()
```

# Visualizar los datos

```python
# Divide the data by response vector
setosa = iris_df[iris_df.Species=='Iris-setosa']
versicolor = iris_df[iris_df.Species=='Iris-versicolor']
virginica = iris_df[iris_df.Species=='Iris-virginica']
```

```python
# Plot the data (axes defined by Google Colab chart suggestions)
plt.title('Iris Data Set')
plt.scatter(setosa.PetalWidthCm, setosa.PetalLengthCm, color='red', label='setosa', alpha=0.5)
plt.scatter(versicolor.PetalWidthCm, versicolor.PetalLengthCm, color='green', label='virginica', alpha=0.5)
plt.scatter(virginica.PetalWidthCm, virginica.PetalLengthCm, color='blue', label='virginica', alpha=0.5)
plt.xlabel('Petal Width')
plt.ylabel('PetalLength')
plt.legend()
```

#Regresión logística

La regresión logística predice el valor de una variable categórica utilizando el valor y la relación entre un conjunto de variables independientes. El algoritmo utiliza las variables independientes para calcular la probabilidad entre 0 y 1 del valor de una variable categórica dependiente utilizando una función logística.

```python
# Normalize/standardize all features by scaling
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
# Train the model using the training set
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
# Make predictions of the testing set using the model
y_pred = lr.predict(x_test)
```

```python
# Calculate the accuracy of the model
print('Exactitud (Regresión logística):', accuracy_score(y_test, y_pred)*100, '%')
```

```python
# Measure the performance of the model
c_m = confusion_matrix(y_test, y_pred)
# Graph the confusion matrix
sns.heatmap(c_m, annot=True, cmap='Blues',
            xticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
            yticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.ylabel('Valor verdadero')
plt.xlabel('Predicción')
plt.title('Matriz de confusión (Regresión logística)')
```

#Bayes

El modelo Naive Bayes calcula individualmente las probabilidades de los valores de la variable categórica para cada variable independiente y combina las probabilidades para determinar el valor de la variable categórica. Es decir, la claificación final es producto de las probabilidades individuales para cada clasificación por cada variable independiente.

```python
# Train the model using the training set
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(x_train, y_train)
```

```python
# Make predictions of the testing set using the model
y_pred = gnb.predict(x_test)
```

```python
# Calculate the accuracy of the model
print('Exactitud (Algoritmo de Bayes):', accuracy_score(y_test, y_pred)*100, '%')
```

```python
# Measure the performance of the model
c_m = confusion_matrix(y_test, y_pred)
# Graph the confusion matrix
sns.heatmap(c_m, annot=True, cmap='Blues',
            xticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
            yticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.ylabel('Valor verdadero')
plt.xlabel('Predicción')
plt.title('Matriz de confusión (Algoritmo de Bayes)')
```

# Análisis de discriminante

El análisis de discriminante proyecta los datos en espacios de menores dimensiones determinados por la o las funciones discriminantes que maximizan la separación entre datos de distintas categorías. Lo anterior permite la categorización de nuevos datos en base a su posición y distancia de las categorías.


```python
# Train the model using the training set
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis()
lda.fit(x_train, y_train)
```

```python
# Make predictions of the testing set using the model
y_pred = lda.predict(x_test)
```

```python
# Calculate the accuracy of the model
print('Exactitud (Análisis de discriminante):', accuracy_score(y_test, y_pred)*100, '%')
```

```python
# Measure the performance of the model
c_m = confusion_matrix(y_test, y_pred)
# Graph the confusion matrix
sns.heatmap(c_m, annot=True, cmap='Blues',
            xticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'],
            yticklabels=['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'])
plt.ylabel('Valor verdadero')
plt.xlabel('Predicción')
plt.title('Matriz de confusión (Análisis de discriminante)')
```
