#Teorema del límite Central (TLC)

Dada una muestra de datos $X_1,X_2,..X_n$ de media $\mu$ y varianza $\sigma^2$ (no importando la distribución), si $n$ es suficientemente grande entonces, $\overline{X}$ ~ $N(\mu,\frac{\sigma^2}{n})$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
lambda_parametro = 0.5
t_muestra = 100000 # numero de elementos por muestra
num_muestras = 1000 # numero de muestras
muestras = np.random.exponential(scale=(1/lambda_parametro), size=(num_muestras, t_muestra))
```

```python
muestras
```

```python
# Calcular el promedio por muestra
medias_muestrales = np.mean(muestras, axis=1)
```

```python
medias_muestrales
```

```python
# Crear un dataFrame para la maedia de las muestras
data = {'Muestra' : range(1,num_muestras+1), 'Media muestral' : medias_muestrales}
df = pd.DataFrame(data)
```

```python
df
```

```python
# Crear un histograma con las medias muestrales
plt.hist(medias_muestrales, bins=30, alpha=0.6, color='b', label='Medias muestrales')
plt.axvline(x=1/lambda_parametro, color='r', linestyle='dashed', linewidth=2, label='mu(µ)')
plt.xlabel('Valores de las medias muestrales')
plt.ylabel('Frecuencia de las medias meustrales (probabilidad)')
plt.title('Teorema del límite central-Distribución exponencial')
plt.legend()
plt.show()
```
